```python
cleaned_Missing_Values = Accidents.dropna(subset=['LSOA_of_Accident_Location'])
cleaned_Missing_Values.count()
```

```python
Cleaned_MV = cleaned_Missing_Values.dropna(subset=['Time'])
```

```python
Location_Cleaned_MV = Cleaned_MV.dropna(subset=['Longitude', 'Latitude', 'Location_Northing_OSGR', 'Location_Easting_OSGR'])
Location_Cleaned_MV.count()
```

```python
# df.dropna(subset=['name', 'born'])
```

```python
# Which type of roads contain the highest number of accidents? (could use google maps api to make heat map of areas 
# with highest amounts of accidents)
# Accident_Index, Road_Type, Longitude, Latitude, Speed_limit, Weather_Conditions, Light_Conditions, Day_of_Week

num_acc_df = Location_Cleaned_MV[['Accident_Index', 'Road_Type','Latitude','Longitude']]
num_acc_df.head()
```

```python
# create a groupby variable of the data by 'road type', and the create a count of accidnts per road type.

Road_Type_count = num_acc_df.groupby('Road_Type')
count_Road_Type = Road_Type_count.count()
count_Road_Type_df = count_Road_Type.drop(['Accident_Index','Latitude'], axis=1)
rename_count_Road_Type_df = count_Road_Type_df.rename(columns={'Longitude':'Number of Accidents'})
rename_count_Road_Type_df
```

```python
# Create a bar plot of the Number of Accidents per Road Type dataset and rename road types according to the road type names.
bar_plot_Road_Type = rename_count_Road_Type_df.plot(kind='bar', title ="Number of Accidents", figsize=(15, 10), legend=True, fontsize=12)

bar_plot_Road_Type.set_ylabel('No. of Accidents')
bar_plot_Road_Type.set_title('Number of Accidents per Road Type')
# bar_plot_Road_Type.set_xticks(ind+width)
bar_plot_Road_Type.set_xticklabels( ('Roundabout', 
                                     'One way street',
                                     'Dual carriageway',
                                     'Single carriageway',
                                     'Slip road',
                                     'Unknown'
                                    ))
bar_plot_Road_Type
```
